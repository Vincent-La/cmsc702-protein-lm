```python
plmdca_file = "../data/cadherin/PLMDCA_apc_fn_scores_Trimmed_PF00028_plmdca.txt"

f = open(plmdca_file, 'r')
```

```python
plmdca_info = []
for line in f:
    if line[0] != '#':
        plmdca_info.append(line.split())

plmdca_info
```

```python
# iterate over plmdca_info and write to csv
plmdca_csv_file = "../data/cadherin/PF00028_plmdca.csv"

with open(plmdca_csv_file, 'w') as plm_csv:
    for data_pt in plmdca_info:
        plm_csv.write("{},{},{}\n".format(data_pt[0], data_pt[1], data_pt[2]))
```
